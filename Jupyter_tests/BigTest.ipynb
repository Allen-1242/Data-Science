```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("b.csv")
df.head()
```

```python
df.info()
```

```python
df1 = df["start_time"]
#df1 = pd.Series.to_frame(df1)
df1.head()
```

```python
df1 = pd.to_datetime(df1)
df1.head()
```

```python
df1_hours = df1.dt.day
df1_hours.head()

```

```python
df1_hours = pd.Series.to_frame(df1_hours)
df1_hours.head()
```

```python
df1_hours.info()
```

```python
df1_hours = df1_hours.rename(columns = {"start_time": "day_taken_out"})
df1_hours.head()
```

```python
df = df.drop("start_time", axis =1)
df.head(1)
```

```python
df.head()
df = pd.concat([df,df1_hours], axis= 1)
df.head(1)
```

```python
df.info()
```
