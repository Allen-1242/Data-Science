```python
df1 = df["start_time"]
#df1 = pd.Series.to_frame(df1)
df1.head()
```

```python
df_f = df_t.value_counts()
df_f.head()
df_f = pd.Series.to_frame(df_f)
df_f.head()
```

```python
df_f = df_f.reset_index()
df_f.head()
```
